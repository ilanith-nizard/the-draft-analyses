# Fitness Is Free

Analysis for [TheDraft](https://open.substack.com/pub/thedrafthq/p/fitness-is-free-but-you-wouldnt-know?r=6hubxb&utm_campaign=post&utm_medium=web&showWelcomeOnShare=true). The piece argues that movement is technically free — running, walking, bodyweight exercise — but the wellness industry has built a stratified market that makes the free version feel inadequate. This notebook documents the scale of that market, the cost spectrum across fitness tiers, and the gap between what people pay and what they actually do.

Data is compiled from the Global Wellness Institute's annual reports, IHRSA gym industry data, contemporary press coverage of boutique fitness (Equinox, SoulCycle, F45, Barry's), and the case-study layer is the author's personal fitness stack. Each price point has a source comment.

```python
# chart 1: global wellness economy and physical activity segment growth, 2015-2024
# establishes the macro scale of the market

fig, ax = plt.subplots(figsize=(10, 6))

color1 = '#2C3E50'
ax.plot(df_wellness['year'], df_wellness['wellness_economy_trillions_usd'],
        marker='o', markersize=10, linewidth=3, color=color1, label='Global Wellness Economy ($T)')
ax.fill_between(df_wellness['year'], df_wellness['wellness_economy_trillions_usd'],
                alpha=0.15, color=color1)

ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Global Wellness Economy (Trillions USD)', fontsize=11, color=color1)
ax.tick_params(axis='y', labelcolor=color1)

ax2 = ax.twinx()
color2 = '#C44536'
ax2.plot(df_wellness['year'], df_wellness['physical_activity_segment_billions_usd'],
         marker='s', markersize=10, linewidth=3, color=color2, linestyle='--',
         label='Physical Activity Segment ($B)')
ax2.set_ylabel('Physical Activity Segment (Billions USD)', fontsize=11, color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

ax.annotate(f"${df_wellness['wellness_economy_trillions_usd'].iloc[-1]}T",
            xy=(2024, df_wellness['wellness_economy_trillions_usd'].iloc[-1]),
            xytext=(-30, 12), textcoords='offset points',
            fontsize=12, fontweight='bold', color=color1)

ax2.annotate(f"${df_wellness['physical_activity_segment_billions_usd'].iloc[-1]:.0f}B",
             xy=(2024, df_wellness['physical_activity_segment_billions_usd'].iloc[-1]),
             xytext=(-35, -18), textcoords='offset points',
             fontsize=12, fontweight='bold', color=color2)

# UPDATED TITLE - drops the misleading "humans haven't become more active" claim
ax.set_title('Global Wellness Economy Growth, 2015-2024',
             fontsize=14, fontweight='bold', pad=15)
ax.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.grid(alpha=0.3)
ax.set_xticks(df_wellness['year'])

lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

plt.tight_layout()
plt.savefig('wellness_economy_growth.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# chart 2: US fitness market segments, color-coded by category
# visualizes that aesthetic + adjacent + equipment dwarf the structural layer

df_segments_sorted = df_segments.sort_values('us_market_billions_usd', ascending=True)

category_colors = {
    'structural': '#2C3E50',
    'equipment':  '#7F8C8D',
    'aesthetic':  '#C44536',
    'adjacent':   '#D4A574',
}

colors = [category_colors[cat] for cat in df_segments_sorted['category']]

fig, ax = plt.subplots(figsize=(11, 7))
bars = ax.barh(df_segments_sorted['segment'], df_segments_sorted['us_market_billions_usd'],
               color=colors, edgecolor='white', linewidth=1.5)

for bar, val in zip(bars, df_segments_sorted['us_market_billions_usd']):
    ax.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2,
            f'${val}B', va='center', fontsize=11, fontweight='bold', color='#333')

ax.set_title('U.S. Fitness Market by Segment\nAthleisure Alone Exceeds Gym + Boutique + Personal Training Combined',
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('Market size (Billions USD)', fontsize=11)
ax.set_xlim(0, max(df_segments_sorted['us_market_billions_usd']) * 1.15)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='x', alpha=0.3)
ax.grid(axis='y', visible=False)

legend_elements = [
    Patch(facecolor='#2C3E50', label='Structural (helps you show up & move)'),
    Patch(facecolor='#7F8C8D', label='Equipment (home replication)'),
    Patch(facecolor='#C44536', label='Aesthetic (signaling)'),
    Patch(facecolor='#D4A574', label='Adjacent (supplements / recovery)'),
]
ax.legend(handles=legend_elements, loc='lower right', fontsize=10, framealpha=0.9)

plt.tight_layout()
plt.savefig('us_fitness_market_breakdown.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# chart 3: fitness cost spectrum — what each tier costs monthly
# from $0 (free outdoor activity) to $1200/month (Tier 5 all-in)
# the author's actual stack (Tier 4) is highlighted

fig, ax = plt.subplots(figsize=(11, 6.5))

# gradient from green (free) to red (most expensive)
tier_colors = ['#27AE60', '#F1C40F', '#E67E22', '#D35400', '#C44536', '#8B0000']

bars = ax.barh(df_tiers['tier'], df_tiers['monthly_cost_usd'],
               color=tier_colors, edgecolor='white', linewidth=1.5)

for bar, monthly, annual in zip(bars, df_tiers['monthly_cost_usd'], df_tiers['annual_cost_usd']):
    width = bar.get_width()
    label = f'${monthly}/mo (${annual:,}/year)' if monthly > 0 else 'Free'
    ax.text(width + 20, bar.get_y() + bar.get_height()/2,
            label, va='center', fontsize=10, fontweight='bold', color='#333')

ax.set_title('The Fitness Cost Spectrum\nWhat Movement Actually Costs Across Tiers',
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('Monthly cost (USD)', fontsize=11)
ax.set_xlim(0, max(df_tiers['monthly_cost_usd']) * 1.4)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='x', alpha=0.3)
ax.grid(axis='y', visible=False)

plt.tight_layout()
plt.savefig('fitness_cost_spectrum.png', bbox_inches='tight', dpi=200)
plt.show()
```

## Key Findings

**The global wellness economy nearly doubled in a decade.** From \$3.7T in 2015 to \$6.8T in 2024 (+84%). The physical activity segment within it grew even faster, from \$542B to $1,143.9B (+111%). Projected to reach \$9.8T by 2029.

**US adult activity has also grown over this period.** The share of US adults meeting CDC aerobic activity guidelines rose from ~26% in the early 2000s to ~47.2% in 2022-2023. Completely sedentary rates dropped from >40% to ~25.6%. So the market growth and the activity growth have moved together.

**But the market has grown disproportionately into signaling layers.** Of the $285B US fitness market:
- **Structural** (gyms + boutique + personal training): \$58B
- **Equipment** (connected fitness + home equipment + apps): \$22B
- **Aesthetic** (athleisure + wearables): \$145B
- **Adjacent** (supplements + recovery): \$60B

For every \\$1 Americans spend on the part of fitness that helps them show up and move (the structural layer), they spend approximately \$3.91 on the surrounding signaling, equipment, and adjacent layers.

**Athleisure alone (\$115B) is larger than the entire structural fitness market combined.**

**The personal cost of fitness spans from \$0/month to \$1,200/month.** Tier 0 (outdoor running, walking, bodyweight) is free. Tier 5 (Equinox + trainer + supplements + recovery + athleisure) is \$14,400/year. The author's stack lands at Tier 4 ($3,815/year), explicitly excluding travel-required races that would push it higher.

**The macro takeaway:** the wellness market's growth isn't fully explained by people exercising more. Activity has grown, but the market has grown into categories that don't directly drive movement — they signal the identity of being a fit person. The free version of fitness still works as well as it ever did. The market has just convinced us it's not enough.
